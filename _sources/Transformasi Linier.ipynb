# TRANSFORMASI

Transformasi linier merupakan sebuah fungsi yang memetakan suatu ruang vector B ke suatu ruang vector C, sehingga variable A dinyatakan sebagai transformasi linier dari B ke C. Pernyataan tersebut ditunjukkan pada skema dibawah ini :

**Jika A : B C**

Variabel akan disebut sebagai transformasi linier jika semua vector u dan v yang ada pada B dan semua scalar c, seperti :

**T(u + v) = T(u) + T(v)**

 **T(cu) = cT(u)**

Transformasi linier akan tampak terlihat jelas jika B = C dan akan dinyatakan
dalam bentuk A: B (B yang disebut dengan operator linier pada

mentranformasikan suatu objek dari dimensi dua ke dimensi lain, diawali dengan satu vektor lalu tranformasi menjadi vektor lain.




Misalkan $u=\binom{u_1}{u_2}$ dan $v=\binom{v_1}{v_2}$
$$
u+v=\binom{u_1}{u_2}+\binom{v_1}{v_2}=\binom{u_1+v_1}{u_2+v_2}
$$

Sekarang, terapkan $T$ pada $u+v$ :
$$
T(u+v)=T\left(\binom{u_1+v_1}{u_2+v_2}\right)=A\binom{u_1+v_1}{u_2+v_2}
$$

Lakukan perkalian matriks:
$$
A\binom{u_1+v_1}{u_2+v_2}=\left(\begin{array}{cc}
-1 & 0 \\
0 & 1
\end{array}\right)\binom{u_1+v_1}{u_2+v_2}=\binom{-\left(u_1+v_1\right)}{u_2+v_2}
$$

Di sisi lain, kita juga hitung $T(u)$ dan $\downarrow v$ ) masing-masing:

$$
\begin{gathered}
T(u)=A\binom{u_1}{u_2}=\left(\begin{array}{cc}
-1 & 0 \\
0 & 1
\end{array}\right)\binom{u_1}{u_2}=\binom{-u_1}{u_2} \\
T(v)=A\binom{v_1}{v_2}=\left(\begin{array}{cc}
-1 & 0 \\
0 & 1
\end{array}\right)\binom{v_1}{v_2}=\binom{-v_1}{v_2}
\end{gathered}
$$

Jumlahkan $T(u)$ dan $T(v)$ :
$$
T(u)+T(v)=\binom{-u_1}{u_2}+\binom{-v_1}{v_2}=\binom{-u_1-v_1}{u_2+v_2}
$$

Sehingga kita melihat bahwa:
$$
T(u+v)=\binom{-\left(u_1+v_1\right)}{u_2+v_2}=\binom{-u_1-v_1}{u_2+v_2}=T(u)+T(v)
$$

 Transformasi $T$ memenuhi sifat aditif $T(u+v)=T(u)+T(v)$.

# CONTOH TRANSFORMASI T = R_{3} ke R_{3}

$T\left(\left[\begin{array}{l}x_1 \\ x_2 \\ x_3\end{array}\right]\right)=\left[\begin{array}{c}x_1+2 x_2+x_3 \\ x_1+5 x_2 \\ x_3\end{array}\right]$

Pengerjaannya :

$\begin{aligned} & \mathbf{e}_1=(1,0,0) \rightarrow \mathrm{T}\left(\mathbf{e}_1\right)=\mathrm{T}\left(\left[\begin{array}{l}1 \\ 0 \\ 0\end{array}\right]\right)=\left(\begin{array}{c}1+2(0)+0 \\ 1+5(0) \\ 0\end{array}\right)=\left(\begin{array}{l}1 \\ 1 \\ 0\end{array}\right) \\ & \mathbf{e}_2=(0,1,0) \rightarrow \mathrm{T}\left(\mathbf{e}_2\right)=\mathrm{T}\left(\left[\begin{array}{l}0 \\ 1 \\ 0\end{array}\right]\right)=\left(\begin{array}{c}0+2(1)+0 \\ 0+5(1) \\ 0\end{array}\right)=\left(\begin{array}{l}2 \\ 5 \\ 0\end{array}\right)\end{aligned}$
$\mathbf{e}_3=(0,0,1) \rightarrow \mathrm{T}\left(\mathbf{e}_3\right)=\mathrm{T}\left(\left[\begin{array}{l}0 \\ 0 \\ 1\end{array}\right]\right)=\left(\begin{array}{c}0+2(0)+1 \\ 0+5(0) \\ 1\end{array}\right)=\left(\begin{array}{l}1 \\ 0 \\ 1\end{array}\right)$

Hasil matriks standard nya

  $$
  A =
  \begin{bmatrix}
  1 & 2 & 1 \\
  1 & 5 & 0 \\
  0 & 0 & 1 \\
  \end{bmatrix}
  $$


# MATRIKS TRANSFORMASI REFLEKSI

Matriks Transformasi Refleksi atau perpindahan matriks yang tampilannya menggunakan metode pencerminan merupakan matriks transformasi yang mempresentasikan operasi refleksi atau pencerminan dari suatu vektor terhadap suatu gari atau bidang yang ada di dalam vektor. dimana pada jenis transformasi linier ini akan memetakan setiap titik pada satu sisi garis atau bidang ke posisi yang sama di sisi lain.



  


  

# CONTOH MATRIKS TRANSFORMASI REFLEKSI

  $$T
  \begin{bmatrix}
  2  \\
  0  \\
  \end{bmatrix}
  =
  \begin{bmatrix}
  -2  \\
  0  \\
  \end{bmatrix}
  $$

  $$T
  \begin{bmatrix}
  2  \\
  2  \\
  \end{bmatrix}
  =
  \begin{bmatrix}
  -2  \\
  2  \\
  \end{bmatrix}
  $$

  Titik (2, 0) direfleksikan menjadi (-2, 0)

  Titik (2, 2) direfleksikan menjadi (-2, 2)

  inisialisasi terlebih dahulu:

  $$
T\left(\binom{x}{y}\right)=A\binom{x}{y}
$$

Diketahui bahwa:
$$
\begin{aligned}
& A\binom{2}{0}=\binom{-2}{0} \\
& A\binom{2}{2}=\binom{-2}{2}
\end{aligned}
$$

Bentuk umum $A$ :
$$
A=\left(\begin{array}{ll}
a & b \\
c & d
\end{array}\right)
$$

Sehingga perhitungan vektor pertamanya:
$$
\left(\begin{array}{ll}
a & b \\
c & d
\end{array}\right)\binom{2}{0}=\binom{2 a}{2 c}=\binom{-2}{0}
$$

Sehingga persamaannya menjadi:
$$
\begin{gathered}
2 a=-2 \Longrightarrow a=-1 \\
2 c=0 \Longrightarrow c=0
\end{gathered}
$$

Perhitungan vektor kedua:
$$
\left(\begin{array}{ll}
a & b \\
c & d
\end{array}\right)\binom{2}{2}=\binom{2 a+2 b}{2 c+2 d}=\binom{-2}{2}
$$

Sehingga persamaannya:
$$
\begin{gathered}
2 a+2 b=-2 \Longrightarrow-2+2 b=-2 \Longrightarrow 2 b=0 \Longrightarrow b=0 \\
2 c+2 d=2 \Longrightarrow 0+2 d=2 \Longrightarrow 2 d=2 \Longrightarrow d=1
\end{gathered}
$$

Jadi, matriks transformasi refleksi $A$ adalah:
$$
A=\downarrow\left(\begin{array}{cc}
-1 & 0 \\
0 & 1
\end{array}\right)
$$
